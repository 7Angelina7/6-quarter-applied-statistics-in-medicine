## Однофакторная ANOVA

Гипотеза: влияет ли уровень ЛПВП-холестерина в крови на размер бляшки в сосуде?

H0 - уровень ЛПВП-холестерина в крови не влияет на размер бляшки в сосуде

H1 - уровень ЛПВП-холестерина в крови влияет на размер бляшки в сосуде

ЗП - уровень ЛПВП-холестерина (HDLi...94), количественная, непрерывная

НП - толщина бляшки (Plaque), категориальная, дискретная

Выборки независимые.

```r
model <- aov(HDLi...94 ~ Plaque, data = data)
summary(model)
```

значимых связей нет => уровень ЛПВП-холестерина в крови не влияет на размер бляшки в сосуде

```r
plot(model)
```

## ANCOVA

НП - Group

ЗП - B_hdl

Ковариата - GPHDL4

```r
model <- aov(B_hdl ~ Group * GPHDL4, data = data)
summary(model)
```

как в группах GPHDL4 влияет на B_hdl?

Group влияет на B_hdl

GPHDL4 влияет на B_hdl

При взаимодействии Group:GPHDL4 влияние нет

```r
plot(model)
```

## MANOVA

ЗП - GPHDL12 - концентрация субфракции липопротеинов

Два фоктора - risk, Plaque

```r
model <- aov(GPHDL12 ~ risk * Plaque, data = data)
summary(model)
```

есть зависимая связь между переменной risk и GPHDL12

post-hoc

```r
TukeyHSD(model)
```

Различия GPHDL12 найдены:
например различия в группе с риском 7 и величиной бляшки 3 и в группе с риском 0 и величиной бляшки 2
