# Эмпирическая гипотеза №1:

влияет ли наличие артериальной гипертензии на поражение артерий атеросклерозом.

Статистические гипотезы:

Н0 - наличие артериальной гипертензии не увеличивает риск образования атеросклероза.

Н1 - наличие артериальной гипертензии увеличивает риск образования атеросклероза.

1 группа: нет артериальной гипертензии (0).

2 группа: есть артериальная гипертензия (1).

Выборки независимые.ЗП - уровень атерогенности. НП - наличие артериальной гипертензии.



```r
ggplot(data, aes(HT)) +
  geom_bar(color="black", fill="lightblue") +
  ggtitle("Распределение наличия артериальной гипертензии") +
  scale_y_continuous(name="частота")
```

## T-test

```r
t.test(log10(data$AI[data$HT == 0]),
       log10(data$AI[data$HT == 1]))
```

уровень значимости высокий, отвергаем нулевую гипотезу, различий нет

```r
# распределене атерогенности по двум группам, отсутствие и наличие артериальной гипертензии
ggplot(data, aes(AI)) +
  geom_histogram(color="black", fill="lightblue") +
  ggtitle("Распределение атерогенности") +
  #scale_x_log10(name="Log10 LDLi") +
  scale_y_continuous(name="частота") +
  theme_minimal() +
  facet_grid(rows = vars(HT))
```

У группы без артериальной гипертензии нормальное распределение, скошено вправо. У группы с артериальной гипертензией распределение с двумя пиками.

```r
ggplot(data, aes(AI)) +
  geom_boxplot(color="black", fill="lightblue") +
  ggtitle("Распределение атерогенности") +
  #scale_x_log10(name="Log10 LDLi") +
  scale_y_continuous(name="частота") +
  theme_minimal() +
  facet_grid(rows = vars(HT))
```

## Тест Манна-Уитни

```r
wilcox.test( data$AI ~ data$HT, data=data, paired = FALSE)
```

# Эмпирическая гипотеза №2:
влияет ли уровень ЛПНП-холестерина в крови на поражение артерий атеросклерозом.

Статистические гипотезы:

Н0 - уровень ЛПНП-холестерина в крови не влияет на поражение артерий атеросклерозом.

Н1 - уровень ЛПНП-холестерина в крови влияет на поражение артерий атеросклерозом.

1 группа: уровень ЛПНП-холестерина (мг/дл) LDLi...95 < 145 мг/дл.

2 группа: уровень ЛПНП-холестерина (мг/дл) LDLi...95 >= 145 мг/дл.

Выборки независимые. ЗП - уровень атерогенности. НП - уровень ЛПНП-холестерина в крови.

```r
ggplot(data, aes(LDLi...95)) +
  geom_histogram(color="black", fill="lightblue") +
  ggtitle("Распределение уровеня ЛПНП-холестерина (мг/дл)") +
  scale_x_log10(name="Log10 LDLi") +
  scale_y_continuous(name="частота")
  theme_minimal
```

распределение стремится к нормальному (необходимо больше данных), скошено вправо.

## T-test

```r
t.test(log10(data$AI[data$LDLi...95 >= 145]),
       log10(data$AI[data$LDLi...95 < 145]))
```

уровень значимости высокий, отвергаем нулевую гипотезу, различий нет

# Эмпирическая гипотеза №3:
зависит ли уровень риска от уровеня наличия атеросклерозных бляшек.

Статистические гипотезы:

Н0 - уровень риска не зависит от уровеня наличия атеросклерозных бляшек.


Н1 - уровень риска зависит от уровеня наличия атеросклерозных бляшек.

1 группа: уровень наличия атеросклерозных бляшек Plaque 1

2 группа: уровень наличия атеросклерозных бляшек Plaque 2

3 группа: уровень наличия атеросклерозных бляшек Plaque 3

Выборки независимые. ЗП - уровень наличия атеросклерозных бляшек. НП - уровень риска.

Колонка риск - это сумма всех десяти факторов риска из списка:

Sm - курение

Dm - диабет

HT - артериальная гипертензия

LVH - левожелудочковая гипертрофия

IHD - ишемическая болезнь сердца

MI - анамнез инфаркта миокарда

Str - анамнез инсульта

A_C - семейный анамнез ИБС

A_H - семейный анамнез артериальной гипертензии

A_D - семейный анамнез диабета
